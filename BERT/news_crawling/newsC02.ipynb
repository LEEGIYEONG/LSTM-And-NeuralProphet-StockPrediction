```python
start = 1
result_df = pd.DataFrame()

while start <= 51:
    try:
        url = 'https://search.naver.com/search.naver?&where=news&query=코스피&sm=tab_pge&sort=0&photo=0&field=0&reporter_article=&pd=3&ds=2021.06.12&de=2021.06.12&docid=&nso=so:r,p:from20210101to20210109,a:all&mynews=0&tart=1&refresh_start=0'.format(start)
        headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36',
          'accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9'}
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'lxml')
        news_title = [title['title'] for title in soup.find_all('a', attrs={'class':'news_tit'})] # 기사 제목
        
        dates = [ date.get_text() for date in soup.find_all('span', attrs={'class':'info'})] # 기사 작성일
        news_date = []
        for date in dates:
            if re.search(r'\d+.\d+.\d+.', date) != None: # 기사 작성일 정제
                news_date.append(date)
        
        df = pd.DataFrame({'date':news_date,'title':news_title})
        result_df = pd.concat([result_df, df], ignore_index=True)
        start += 10
        print(start)
    
    except: # 오류발생시 몇 페이지까지 크롤링했는지 page를 확인하기 
        print(start)
        break
```

```python
result_df["title"] = result_df["title"].str.replace(pat=r'[^\w]', repl=r'', regex=True)
result_df["title"] = result_df["title"].str.replace(pat=r'[A-Z]', repl=r'', regex=True)
#result_df["title"] = result_df["title"].str.replace(pat=r'[0-9]', repl=r'', regex=True)
```
